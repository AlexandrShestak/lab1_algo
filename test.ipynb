```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (20, 5)
```

```python
n = []
timeSimple = []
timeBlock = []
timeSwapped = []
with open('result.txt') as f:
    lines = f.readlines()
    for line in lines:
        n.append(int(line.split()[0]))
        timeSimple.append(line.split()[1])
        timeBlock.append(line.split()[2])
        timeSwapped.append(line.split()[3])
plt.plot(n, timeSimple, label='timeSimple', color='black')
plt.plot(n, timeBlock, label='timeBlock', color='red')
plt.plot(n, timeSwapped, label='timeSwapped', color='green')
plt.legend()
plt.xticks(np.arange(1000, 2000, 50))

plt.show()
```
